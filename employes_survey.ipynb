### 1. How satisfied are employees with their current roles based on survey responses? 

```python
df = pd.read_csv("employee_survey.csv")
df.head(10)
```

```python
print("average empoyeees job satisfaction  :",df["JobSatisfaction"].mean())
```

```python
df["JobSatisfaction"].value_counts()
```

###  3. What is the average job satisfaction score for each department?

```python
df.groupby("Dept")["JobSatisfaction"].agg("mean") 
```

###  4. How many employees are in each age group?

```python
pd.cut(df["Age"],bins =[0,25,35,45,70],labels =["young","middle age","adult","old"]).value_counts()
```

###  5. What is the average years of service for each education level? 


```python
round(df.groupby("EduLevel")["Experience"].mean(),1)
```

### 6.Which departments have the highest and lowest employee engagement scores? 

```python
dept_engagement = df.groupby("Dept")["WorkEnv"].mean().sort_values(ascending=False)


highest_engagement = dept_engagement.idxmax(), dept_engagement.max()
lowest_engagement = dept_engagement.idxmin(), dept_engagement.min()


print("\nHighest Engagement Department:", highest_engagement)
print("Lowest Engagement Department:", lowest_engagement)
```

### 7. How many employees in each department have a "Very Satisfied" job satisfaction score?

```python
df[df["JobSatisfaction"] == 5].groupby("Dept")["EmpID"].count()
```

###  8. What is the average workload for each job role?
